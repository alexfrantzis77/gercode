```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the chart
categories = [1, 2, 3, 4, 5]
counts = np.array([2, 2, 2, 5, 13])
percentages = np.round(counts / sum(counts) * 100, 1)

# Plotting
plt.figure(figsize=(6, 3), dpi=300)
plt.bar(categories, counts, color='darkblue', zorder=0)

# Adding labels on top of the bars
for i, (value, percentage) in enumerate(zip(counts, percentages)):
    plt.text(categories[i], value - 1, f"{value} ({percentage}%)", ha='center', fontsize=8, color='white', weight='bold', zorder=10)

# Customizing the plot
xticks = ['1\nNot helpful at all', 2, 3, 4, '5\nExtremely helpful']
plt.xticks(categories, xticks)
plt.yticks(range(0, 16, 5))
plt.ylabel('Count')
#plt.title('How helpful was the programming assignment in improving\nyour understanding of soil classification concepts?\n', weight='bold')
plt.ylim(0, 15)
plt.savefig('Q1.png', bbox_inches='tight')
plt.show()
```
